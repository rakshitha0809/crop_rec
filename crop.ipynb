```python
import pandas as pd
```

```python
df=pd.read_csv('Crop_recommendation.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
x=df.drop('label',axis=1)
y=df['label']
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2,train_size=0.8)
```

```python
x.info()
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
y_train.info()
```

```python
y_test.info()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred1=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
logistic_reg_acc=accuracy_score(y_test,y_pred1)
```

```python
print("accuracy using logistic Regression:"+str(logistic_reg_acc))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier()
```

```python
model2.fit(x_train,y_train)
```

```python
y_pred3=model2.predict(x_test)
```

```python
DecisionTree_reg_acc=accuracy_score(y_test,y_pred3)
print("accuracy using DecisionTree Regression:"+str(DecisionTree_reg_acc))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier()
```

```python
model3.fit(x_train,y_train)
```

```python
y_pred4=model3.predict(x_test)
```

```python
random_acc=accuracy_score(y_test,y_pred4)
print("accuracy using Random Forest Classification:"+str(random_acc))
```
